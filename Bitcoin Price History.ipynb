```python
############################
# Import Packages
############################



import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
############################
# Load Data
############################


btc_data = pd.read_csv('BTC-USD.csv')
btc_data['Datetime'] = pd.to_datetime(btc_data['Date'])
btc_data.head()
```

```python
############################
# BTC Line Plot
############################

ax = sns.lineplot(data = btc_data, x = 'Date', y = 'Close')
ax.xaxis.set_major_locator(ticker.MultipleLocator(365))
plt.xticks(rotation = 45)
plt.title('BTC Price Over Time')
plt.show()
```

```python
############################
# BTC Semilog Plot
############################

ax = sns.lineplot(data = btc_data, x = 'Date', y = 'Close')
ax.xaxis.set_major_locator(ticker.MultipleLocator(365))
ax.set_yscale('log')
plt.xticks(rotation = 45)
plt.title('BTC Price Over Time')
plt.show()


"""
Increase of bitcoin price is more linear on a log scale
"""
```
